```python
def level(count):
    if count==10:
        print('==================================================')
        print()
        print('You are upgraded to "Bronze level!"')
        print()
        print('==================================================')
    elif count==20:
        print('==================================================')
        print()
        print('You are upgraded to Silver level!')
        print()
        print('==================================================')
    elif count==30:
        print('==================================================')
        print()
        print('You are upgraded to Gold level!')
        print()
        print('==================================================')
    elif count==40:
        print('==================================================')
        print()
        print('You are upgraded to Platinum level!')
        print()
        print('==================================================')
    elif count==50:
        print('==================================================')
        print()
        print('You are upgraded to Diamond level!')
        print()
        print('==================================================')
    elif count==60:
        print('==================================================')
        print()
        print('You are upgraded to Ace level!')
        print('You are in Top 1%.')
        print()
        print('==================================================')




import random
summ=0
count = 0

print('==================================================')
print()
print('Welcome to MathMind!!!')
print()
print('You have to choose from 2 cards. Sum of your card and recent value must be between 0 and 21. Initially recent value is 0. If your value not fall in the range 0-21 then game is ended.')
print()
print('==================================================')

while(summ>=0 and summ<=21):
    num1 = random.randint(-9,15)
    num2 = random.randint(-9,15)
    print('card 1: ',num1)
    print('card 2: ',num2)
    print('Your recent value is: ',summ)
    choice = int(input('Enter which card you choose, Card 1 or Card 2(enter 1 or 2): '))
    if choice==1:
        summ+=num1
    elif choice==2:
        summ+=num2
    else:
        print('==================================================')
        print()
        print('You have entered wrong card number.')
        print()
        print('==================================================')
        break
        
    count+=1
    
    level(count)

print('==================================================')
print()
print('You played well! . Game Ended')
print()
print('==================================================')
```
